## Bivariate KDEplot by Category

```python
cars.head()
```

```python
cyl_4 = cars[cars.cylinders==4]
cyl_8 = cars[cars.cylinders ==8]

plt.figure(figsize=(8, 6))

sns.kdeplot(cyl_4, x = "horsepower", y = "mpg",
            cmap = "Blues", shade = True, shade_lowest = False) #cmap = colormap 
sns.kdeplot(cyl_8, x = "horsepower", y = "mpg",
            cmap = "Blues", shade = True, shade_lowest = False)

plt.annotate("4 Cylinders", (105, 32), color = "blue", fontsize = 16, fontweight = 'bold')
plt.annotate("8 Cylinders", (190, 18), color = "blue", fontsize = 16, fontweight = 'bold')
```
