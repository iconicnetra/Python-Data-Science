## While Looping

```python
a = 0 #Initialization

while a < 5: #Print from 0 to 4
    print(a)
    a += 1
```

```python
a = 0 #Initialization

while a < 5: #Print from 0 to 4
    print(a)
    a = a + 1 #Alternative Way to write a += 1
```

```python
i = 0

while i < 50:
    print(i)
    i += 1
```

### The break statements

```python
i = 1
while i <= 50: 
    print(i)
    i += 1
    if i == 10:
        break
```
