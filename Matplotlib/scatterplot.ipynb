### From CSV File

```python
data = pd.read_csv("E:\Python\Python Data Science\Matplotlib\data.csv")
data.head()
```

```python
#Math_Marks Vs. CS Marks 

plt.style.use('fivethirtyeight')

Math_Marks = data['Marks_Math']
CS_Marks = data['CS_Marks']

plt.scatter(Math_Marks, CS_Marks, marker='o', c = 'g', label = "Math Vs. CS")

plt.xlabel("Marks in Mathematics")
plt.ylabel("Marks in CS")
plt.title("Relation Plot of Math Marks VS. CS Marks", fontsize = 15, c = 'r')

plt.legend()
plt.show()
```

### adding the Regression line

```python
import numpy as np 

#Calculating the parameters for the regression line
m, b = np.polyfit(Math_Marks, CS_Marks, 1)
m, b
```

```python
plt.style.use('fivethirtyeight')

Math_Marks = data['Marks_Math']
CS_Marks = data['CS_Marks']

plt.scatter(Math_Marks, CS_Marks, marker='o', c = 'g', label = "Math Vs. CS")

#Plotting the Regression Line 
plt.plot(Math_Marks, m*Math_Marks+b, color = 'red', linewidth = 0.7)

plt.xlabel("Marks in Mathematics")
plt.ylabel("Marks in CS")
plt.title("Relation Plot of Math Marks VS. CS Marks", fontsize = 15, c = 'r')

plt.legend()
plt.show()
```
