# ScatterPlot

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Loading dataset

diamonds = sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds.describe()
```

```python
diamonds.tail()
```

#### Filtering Data

```python
diamond = diamonds[diamonds.cut.isin(['Premium', 'Good']) & 
                    diamonds.color.isin(['D', 'F', 'J'])].sample(n = 100, random_state=22)
diamond.shape
```

```python
sns.set_style("dark")
```

```python
sns.scatterplot(diamonds, x = 'carat', y = 'price', color = 'k', s = 20)
 #s = Size of the plot 
plt.xlabel("Carat")
plt.ylabel("Price")

plt.show()
```

```python
sns.scatterplot(diamond, x = 'x', y = 'y', color = 'k', s = 20)

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

## Plotting Multiple scatter Plot

```python
### Semantic Variables 
sns.set_style("darkgrid")
```

```python
sns.scatterplot(x = 'carat', y = 'price', data=diamond,
                hue = 'cut')
```

```python
sns.scatterplot(data=diamonds,
                x = 'carat', y = 'price', hue = 'cut',
                palette=['purple', '#55CCCC'])

plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Carat Vs. Price Scatter PLot")

plt.show()
```

```python
diamond = diamonds[(diamonds.cut.isin(['Premium', 'Good'])) & 
            (diamonds.color.isin(['D', 'F', 'J']))
            ].sample(n=100, random_state=22)
```

```python
diamond.shape
```

```python
diamonds.shape
```

```python
sns.scatterplot(data=diamond,
                x = 'carat', y = 'price', hue = 'cut',
                palette=['purple', '#55CCCC'])

plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Carat Vs. Price Scatter PLot")

plt.show()
```

```python
## Continuous Variable 
sns.scatterplot(data=diamond, x = 'carat', y = 'price', hue = 'depth')
```

```python
sns.scatterplot(x='carat', y='price', size='cut',
                data=diamond, 
                sizes=[150, 50, 30, 60, 70]
               );
```

### Using Style 

```python
sns.scatterplot(data=diamond, x = 'carat', y = 'price',
                hue = 'cut', style='cut', s = 100)
```

```python
sns.scatterplot(data=diamond, x = 'carat', y = 'price',
                hue = 'cut', style='color', s = 20)
```

```python
sns.scatterplot(data=diamond, x = 'carat', y = 'price',
                hue = 'cut', style='color', size = 'depth')
```

```python
sns.scatterplot(x='carat', y='price', 
                data=diamond, 
                s=200,
                marker='*'
               );
```

```python
sns.scatterplot(x='carat', y='price', 
                data=diamond, 
                s=50, 
                marker='*',
                edgecolor='black'
               );
```
