## Importing Libraries

```python
import numpy as np 
import pandas as pd 
%matplotlib inline
import matplotlib as mpl 
import matplotlib.pyplot as plt 
```

### Importing Data From the Web

```python
data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv")
data.head()
```

```python
data.head(3)
```

```python
print(data.shape)
```

### Seting the Indexy to Country

```python
data.set_index("Country", inplace=True) #It removes the original index and keeps the country as index values 

data.head()
```

```python
### Printing the Dimensions 

print("Data Dimensions:", data.shape)
```

```python
#Finally Creating a List of Years 
#Making easy for plotting the data

years = list(map(str, range(1980, 2014)))
#Example: years = np.arange(1990, 2014)
years
```

# Line Plot

```python
data_line = data[years] #For keeping into data frame to plot the line 
data_line
```

```python
#Getting sum of Immigrants 

total_immigrants = data_line.sum()
total_immigrants
```

```python
#Creating the Figures with Axes 

fig, ax = plt.subplots()

ax.plot(total_immigrants)

#Setting up the Titles 
ax.set_title("Immigrants between 1980 to 2013")

#Setting Up the Labels
ax.set_xlabel("Years")
ax.set_ylabel("Total Immigrants")

#Setting up Legends
ax.legend()

#Display the Plot
plt.show()
```

#### There is issues in X-Labels...to fix this, we need to change the Index Value

```python
#Creating the Figures with Axes 

fig, ax = plt.subplots()

#Changing Index type to Integer
total_immigrants.index = total_immigrants.index.map(int)

ax.plot(total_immigrants)

#Setting up the Titles 
ax.set_title("Immigrants between 1980 to 2013")

#Setting Up the Labels
ax.set_xlabel("Years")
ax.set_ylabel("Total Immigrants")

#Setting up Legends
ax.legend()

#Display the Plot
plt.show()
```

#### Adding Grid, Legend, Limits

```python
#Creating the Figures with Axes 

fig, ax = plt.subplots()

#Changing Index type to Integer
total_immigrants.index = total_immigrants.index.map(int)

ax.plot(total_immigrants, 
        marker = 's', markersize = 5, color = 'green',
        linestyle = 'dashed', dashes = (5, 2)) #Custom Dashed

#Setting up the Titles 
ax.set_title("Immigrants between 1980 to 2013")

#Setting Up the Labels
ax.set_xlabel("Years")
ax.set_ylabel("Total Immigrants")

#Limits on X-axis
plt.xlim(1975, 2015)

#Enabling Grid
plt.grid(True)

#Legend
plt.legend(['Immigrants'])

#Display the Plot
plt.show()
```

```python

haiti = data[data['Country']=="Haiti"]

#Transposing to get the result as series
haiti = haiti[years].T

#Converting the Index
haiti.index = haiti.index.map(int)

#Plotting the Line Plot 
fig, ax = plt.subplots()
ax.plot(haiti)

#Setting Up the Titles and Labels
ax.set_title("Immigrants from Haiti")
ax.set_xlabel("Years")
ax.set_ylabel("# of Immigrants")

plt.legend(['Immingrants'])
plt.grid(True)

plt.show()
```

```python
#Enabling Ticks

haiti = data[data['Country']=="Haiti"]

#Transposing to get the result as series
haiti = haiti[years].T

#Converting the Index
haiti.index = haiti.index.map(int)

#Plotting the Line Plot 
fig, ax = plt.subplots()
ax.plot(haiti)

#Setting Up the Titles and Labels
ax.set_title("Immigrants from Haiti")
ax.set_xlabel("Years")
ax.set_ylabel("# of Immigrants")
#Enabling Ticks
ax.set_xticks(list(range(1980, 2015, 5)))

plt.legend(['Immingrants'])
plt.grid(True)

plt.show()
```
