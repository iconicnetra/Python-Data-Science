# Seaborn Line Plot 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00482/dataset.zip"

data = pd.read_csv(url, parse_dates=[3])
data.head(10)
```

## Cleaning Some Data

```python
#Renaming some data 
data.rename(columns={'SystemCodeNumber': 'Location', 'LastUpdated': 'Timestamp'}, inplace=True)

# Convert 'Timestamp' column to datetime
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Extract additional features
data['Day'] = data['Timestamp'].dt.date
data['Month'] = data['Timestamp'].dt.month
data['Hour'] = data['Timestamp'].dt.hour
 
```

```python
url = "E:\Python\Data\data.csv"

df = pd.read_csv(url)
df.head()
```

```python
#Renaming some of the coloumns
df.rename(columns={"Gender":"gender", "Marks_Math":"math-marks", "CS_Marks":"cs-marks", "Religion":"religion"}, inplace=True)
df.head(10)
```

### Plotting Lines 

```python
plt.figure(figsize=(10, 8))

sns.lineplot(data = df, x = "math-marks", y  = 'cs-marks', color = 'green').set(xlabel = "Marks in Math", ylabel = "Marks is CS",
                                                               xlim = (30, 90), ylim = (30, 50),
                                                               title = "Line Plot of Math Vs. CS Marks")
```

```python
plt.figure(figsize=(10, 8))

sns.lineplot(data = df, x = "math-marks", y  = 'cs-marks', 
             color = 'green',
             hue='gender', style='gender', 
             weights = True).set(xlabel = "Marks in Math", ylabel = "Marks is CS",
                                                               xlim = (30, 90), ylim = (30, 50),
                                                               title = "Line Plot of Math Vs. CS Marks")
```

```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the theme and figure size
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12, 8))

# Create some sample data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.2, size=100)
hue = np.random.choice(["A", "B"], size=100)
size = np.random.choice([1, 2, 3], size=100)
style = np.random.choice(["o", "x"], size=100)
units = np.random.choice(["U1", "U2"], size=100)
weights = np.random.uniform(0.5, 1.5, size=100)

# Create a dataframe from the data
df_1 = pd.DataFrame({"x": x, "y": y, "hue": hue, "size": size, "style": style, "units": units, "weights": weights})

# Draw the line plot with all the parameters and options
sns.lineplot(data=df_1, x="x", y="y", hue="hue", size="size", style="style", units="units", weights="weights",
             palette="Set2", hue_order=["B", "A"], hue_norm=(0, 1), sizes=(1, 5), size_order=[3, 2, 1], size_norm=(0, 10),
             dashes=False, markers=True, style_order=["x", "o"], estimator="mean", errorbar=('ci', 95), n_boot=1000, seed=42,
             orient="x", sort=True, err_style="bars", err_kws={"capsize": 5}, legend="brief", ci=95)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data = df, x = 'math-marks', y = 'cs-marks', hue='religion',
             style='religion', dashes=False, markers=True, estimator='mean', errorbar=('ci', 90))

plt.xlabel("Marks in Math")
plt.ylabel("Marks in Science")

plt.legend(loc = 1)

plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data = df, x = 'math-marks', y = 'cs-marks', hue='religion',
             style='religion', dashes=False, markers=True, estimator='mean', errorbar=None)
sns.despine(left=True)
plt.xlabel("Marks in Math")
plt.ylabel("Marks in Science")

plt.legend(loc = 1)

plt.tight_layout()

plt.show()
```
