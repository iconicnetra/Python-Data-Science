```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Intro Visual 

```python
plt.rc('xtick', labelsize = 14)
plt.rc('ytick', labelsize = 14)
```

```python
xlabels = ['Mon', 'Tues', 'Wed', 'Thus', 'Fri']
ylabels = ['Cofee', 'Juice', 'Soda', 'Tea']

orders = np.array([[44, 31, 50, 41, 30],
                   [21, 12, 28, 38, 35],
                   [15, 18, 20, 14, 12],
                   [38, 29, 39, 26, 11]])
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(orders, 
            cmap='YlOrBr',
            vmin=0,
            xticklabels=xlabels,
            yticklabels=ylabels,
            annot=True,
            alpha=0,
            square=True,
            annot_kws={'fontsize':14, 'fontweight': 'bold', 'color': 'black'}
           )
plt.yticks(rotation=0)
plt.tick_params(
    which='both',      
    bottom=False,      
    left=False,      
    labelbottom=False,
    labeltop=True) 
plt.tight_layout();
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(orders, 
            cmap='YlOrBr',
            vmin=0,
            xticklabels=xlabels,
            yticklabels=ylabels,
            annot=True,
            square=True,
            annot_kws={'fontsize':14, 'fontweight': 'bold', 'color': 'black'}
           )
plt.yticks(rotation=0)
plt.tick_params(
    which='both',      
    bottom=False,      
    left=False,      
    labelbottom=False,
    labeltop=True) 
plt.tight_layout();
```

# Heatmap Graphs 

```python
cars = sns.load_dataset('mpg')
cars.head()
```

```python
cars.shape
```

```python
cars.describe()
```

```python
#Group the element
cars.groupby('origin').cylinders.value_counts()
```

```python
#Converting into dataframe
cars.groupby('origin').cylinders.value_counts().unstack()
```

```python
#Dropping or removing NaN by 0
cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
```

```python
#Keeping into one dataframe
origin_cyl = cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
origin_cyl
```

```python
#Creating Heatmap
sns.heatmap(origin_cyl)
```
