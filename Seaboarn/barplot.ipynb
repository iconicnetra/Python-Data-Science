# Seaborn Barplot

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set_style("darkgrid")
```

```python
#Loading Dataset
penguins = sns.load_dataset("penguins")
penguins.shape
```

```python
penguins.head()
```

# Basics

```python
sns.barplot(penguins, x = 'species', y = 'body_mass_g')
```

```python
#Using hue to add colors
sns.barplot(penguins, x = 'species', y = 'body_mass_g', hue = 'species') # Note: Same as x
plt.title('Barplot of species based on Body Mass')
```

### Finding the Mean Value 

```python
penguins.groupby('species').body_mass_g.mean()
```

#### Horizontal BarPlot 

```python
sns.barplot(penguins, y = 'species', x = 'body_mass_g', hue = 'species') #Interchange x and y
```

## Interval Estimate
Confidence Interval (CI)

```python
sns.barplot(penguins, x = 'species', y = 'body_mass_g', 
            errorbar= ('ci', 90)) 
```

```python
sns.barplot(penguins, x = 'species', y = 'body_mass_g', 
            errorbar= ('ci', 0)) # 0 for None
```

```python
sns.barplot(penguins, x = 'species', y = 'body_mass_g', hue = 'species', 
            errorbar= None) #Next Method
```

### Estimator

```python
import numpy as np
```

```python
penguins.groupby('species').body_mass_g.std()
```

```python
sns.barplot(penguins, x = 'species', y = 'body_mass_g', 
            hue= 'species', estimator = np.std) 
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(penguins, x = 'species', y = 'body_mass_g', 
            hue= 'species', estimator = np.max) 
```

```python
sns.barplot(penguins, x = 'species', y = 'body_mass_g', 
            hue= 'species', estimator = np.mean) 
```

# Stacked Barplot 

```python
penguins_sample = penguins.dropna().groupby(['species', 'sex']).sample(3, random_state=10)
penguins_sample
```

```python
sns.barplot(data=penguins_sample[penguins_sample.sex == 'Female'], 
            x = 'species', y = 'body_mass_g',
            color = 'tab:blue', 
            order = ['Adelie', 'Chinstrap', 'Gentoo'],
            errorbar=None, 
            estimator=np.sum,
            saturation=1
            )
sns.barplot(data=penguins_sample[penguins_sample.sex == 'Male'], 
            x = 'species', y = 'body_mass_g',
            color = 'tab:cyan', 
            order = ['Adelie', 'Chinstrap', 'Gentoo'],
            errorbar=None, 
            estimator=np.sum,
            bottom = (
                penguins_sample[penguins_sample.sex == 'Female']
                .groupby('species')
                .body_mass_g
                .sum()
            ), saturation=1
            )
```

# Another Easy Method

```python
(penguins_sample
    .groupby(['species', 'sex'])
    .body_mass_g
    .sum()
    .unstack()
).plot(kind='bar', stacked=True, color=['tab:blue', 'tab:cyan'], width=0.8)

plt.xticks(rotation = 0)
plt.ylabel("Sum of body Mass (g)")
```
