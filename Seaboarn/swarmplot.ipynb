```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Swarm Plot 

```python
#Loading Dataset 
cars = sns.load_dataset("mpg").dropna()
cars.info()
```

```python
cars.head()
```

```python
cars.tail()
```

```python
cars.describe()
```

## Bascis

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

```python
#Filtering Data
cars = cars[cars.cylinders.isin([4, 6, 8])] #Selecting only key values for the plotting purposes
cars.head()
```

```python
sns.swarmplot(cars.horsepower)
```

### Multiple Swarmplot

```python
#Method: sns.swarmplot(x="category_column", y="numeric_column", data=data)
sns.swarmplot(data=cars, x = "origin", y = "horsepower", hue = "origin")

plt.legend(loc = 3)
```

```python
cars.head(1)
```

```python
sns.swarmplot(data=cars, x = "cylinders", y = "horsepower", hue = "cylinders")
```

```python
sns.swarmplot(data=cars,
              x = 'origin', y = "horsepower",
              hue='origin')
```

```python
#Using Different hue option 
sns.swarmplot(data=cars,
              x = 'origin', y = "horsepower",
              hue='cylinders')
```

```python
#Using Different hue option 
sns.swarmplot(data=cars,
              x = 'origin', y = "horsepower",
              hue='cylinders', 
              dodge=True) #Putting elements side by side
```

### Swarmplot Overlay

```python
usa = cars[cars.origin=='usa']
usa.head()
```

```python
usa.shape
```

```python
cars.origin.value_counts()
```

## With Boxplot

```python
sns.boxplot(usa, x = "cylinders", y = "horsepower")
sns.swarmplot(usa, x = "cylinders", y = "horsepower")
```

```python
sns.swarmplot(usa, x = "cylinders", y = "horsepower")
sns.boxplot(usa, x = "cylinders", y = "horsepower")
```

```python
import numpy as np
sns.boxplot(data=usa, x = "cylinders", y = "horsepower",
              whis = np.inf)
sns.swarmplot(data=usa, x = "cylinders", y = "horsepower",
            color='black', alpha = 0.5)
```

### With Violin Plot 

```python
sns.violinplot(data=usa, x = "cylinders", y = 'horsepower')

sns.swarmplot(data=usa, x = 'cylinders', y = 'horsepower', color = 'black', alpha = 0.5)
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               scale='width'
              )
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='white', size=4
             );
```

```python
sns.violinplot(data=usa, x = "cylinders", y = 'horsepower',
               density_norm='width', inner=None, color = 'xkcd:forest green')
#Density norms are: area, count, width

sns.swarmplot(data=usa, x = 'cylinders', y = 'horsepower', 
              color = 'black', alpha = 0.5, size = 3)
```
