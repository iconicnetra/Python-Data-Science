# Seaborn JointGrid 

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
cars = sns.load_dataset('mpg')
cars.head()
```

## Basics 

```python
sns.set_style('darkgrid')
```

```python
g = sns.JointGrid()
```

```python
type(g)
```

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
```

## `.plot()`

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
g.plot(sns.scatterplot, sns.histplot, color = 'indigo')
```

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
g.plot(sns.regplot, sns.kdeplot, color = 'indigo')
```

# JointPlot and Marginal Plot 

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
g.plot_joint(sns.regplot, color = 'indigo')
g.plot_marginals(sns.kdeplot, fill=True, alpha = 0.5, color = 'green')
```

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
g.plot_joint(sns.regplot, color = 'indigo')
g.plot_marginals(sns.kdeplot, fill=True, alpha = 0.5)
plt.title('JointPlot')
```

## JointGrid Axes

```python
g = sns.JointGrid()
```

```python
type(g.ax_marg_x)
```

```python
g = sns.JointGrid()
x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint)
sns.kdeplot(x=x, y=y, ax=g.ax_marg_x)
sns.histplot(y=y, ax=g.ax_marg_y)
```

## Working with Hue 

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg', hue = 'origin')
g.plot(sns.scatterplot, sns.kdeplot)
```

### Separating according to Axes 

```python
g = sns.JointGrid()

x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax = g.ax_joint)
sns.kdeplot(x=x, y=y, ax = g.ax_marg_x, hue=cars.origin, bw_adjust=10)
sns.histplot(y=y, ax = g.ax_marg_y)
```

```python
g = sns.JointGrid()

x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax = g.ax_joint)
sns.kdeplot(x=x, y=y, ax = g.ax_marg_x, hue=cars.origin, bw_adjust=10)
sns.histplot(y=y, ax = g.ax_marg_y)
```

## Styling 

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
g.plot(sns.scatterplot, sns.histplot, alpha = 0.8, edgecolor = 'black')
```

## Plot Specifice Keywords

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
g.plot_joint(sns.regplot, line_kws = {'color':'black'})
g.plot_marginals(sns.kdeplot, fill=True, color='purple')
```

## JointGrid Properties

```python
g = sns.JointGrid(cars, x = 'weight', y = 'mpg')
g.plot(sns.scatterplot, sns.histplot)
g.set_axis_labels(xlabel="Weight (lbs)", ylabel="Miles per Gallon", fontsize = 20)
plt.title("Weight Vs. Miles")
```

## Thumbnail and Overlays 

```python
g = sns.JointGrid()
ax = g.ax_joint
#Removing tick labels 
ax.set_xticklabels([])
ax.set_yticklabels([])
```

```python
g = sns.JointGrid()
x, y = cars.weight, cars.mpg
sns.scatterplot(x=x, y=y, ax=g.ax_joint, s=60, alpha=0.8, ec='gray')
sns.histplot(x=x, ax=g.ax_marg_x, bins=20)
sns.kdeplot(y=y, ax=g.ax_marg_y, fill=True, lw=3)
g.set_axis_labels(xlabel='', ylabel='')

ax = g.ax_joint
ax.set_xticklabels([])
ax.set_yticklabels([]);
#g.savefig('25jointgrid_image_differentPlots.png', dpi=200);
```

## Thumbnail

```python
sns.set_style('white')

g = sns.JointGrid()
x,y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint, ci=99, 
            scatter_kws={'color': 'xkcd:lavender', 'alpha': 0.7, 'ec': 'gray', 's': 50}, 
            line_kws={'color': 'black', 'lw': 3}
           )
sns.kdeplot(x=x, ax=g.ax_marg_x, fill=True, color='xkcd:indigo', lw=2 )
sns.histplot(y=y, bins=15, ax=g.ax_marg_y, lw=2, color='xkcd:indigo', alpha=0.5, ec='xkcd:indigo')
g.set_axis_labels(xlabel='', ylabel='')

ax = g.ax_joint
ax.set_xticklabels([])
ax.set_yticklabels([]);
#g.savefig('25jointgrid_image_thumb.png', dpi=200);
```
