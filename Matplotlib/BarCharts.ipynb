```python
import matplotlib.pyplot as plt 
```

```python
import csv
import numpy as np 
from collections import Counter
```

```python
#using Styles 
plt.style.use("ggplot")

religions = []

#Reading data from CSV file

with open('data.csv', 'r') as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        religions.append(row['Religion'])

#Counting the occurrence in the Religion Column
religion_counts = Counter(religions)

#Extract unique religions and their corresponding counts 
unique_religions = list(religion_counts.keys())
counts = list(religion_counts.values())

#Creating a Bar Plot 
plt.bar(unique_religions, counts, color = 'g', label = "Religion")


plt.show()
```

```python
#Creating Lables 

#using Styles 
plt.style.use("ggplot")

religions = []

#Reading data from CSV file

with open('data.csv', 'r') as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        religions.append(row['Religion'])

#Counting the occurrence in the Religion Column
religion_counts = Counter(religions)

#Extract unique religions and their corresponding counts 
unique_religions = list(religion_counts.keys())
counts = list(religion_counts.values())

#Creating a Bar Plot 
plt.bar(unique_religions, counts, color = 'g', label = "Religion")
#--------------------------------------------
#Adding Labels and Titles 
plt.xlabel("Religion")
plt.ylabel("Number of Respondents")
plt.title("Distribtuion of Religion")

plt.grid(True, color = 'k', alpha = 0.3)
plt.legend()

plt.show()
```
