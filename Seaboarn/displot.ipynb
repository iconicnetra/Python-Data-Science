# Seaborn Displot

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.head()
```

# Basics 

```python
sns.set_style('white')
```

```python
sns.displot(cars, x = 'weight')
```

```python
g = sns.displot(cars, x = 'weight', color = 'indigo')
plt.title('Histogram')
```

## KDE Plot

```python
sns.displot(cars, x = 'weight', kde = True)
```

## Kind 

```python
sns.displot(cars, x = 'weight', kind = 'kde')
```

```python
sns.displot(cars, x = 'weight', kind = 'hist', color = 'indigo')
```

```python
sns.displot(cars, x = 'weight', kind = 'ecdf', color='indigo')
```

### Adding Rug 

```python
sns.displot(cars, x = 'weight', kind = 'kde', rug=True)
```

## Categorical Variables 

```python
sns.displot(cars, x = 'weight', hue = 'origin')
```

```python
sns.histplot(cars, x = 'weight', hue = 'origin')
```

```python
sns.displot(cars, x = 'weight', hue = 'origin', kind = 'kde')
```

## Bivariate (2D)

```python
sns.histplot(cars, x = 'weight', y = 'horsepower')
```

```python
sns.displot(cars, x = 'weight', y = 'horsepower', kind = 'kde')
```

```python
sns.displot(cars, x = 'weight', y = 'horsepower', kind = 'kde', hue='origin')
```

```python
sns.displot(x='weight', y='horsepower', data=cars, kind='kde', rug=True);
```

## Small Multiples (FacetGrid)

```python
sns.displot(cars, x = 'weight', hue='origin', col='origin')
```

```python
sns.displot(cars, x = 'weight', hue='origin', col='origin', kind = 'kde')
```

### Adding Rows and columsn -- Multiple Graphs 

```python
sns.displot(cars[cars.cylinders.isin([4, 6, 8])], x = 'weight', 
            hue='cylinders', 
            col='origin',
            row='cylinders')
```

## Styling 

```python
sns.displot(cars, x = 'weight', 
            hue = 'origin',
            col = 'origin',
            palette='crest')
```

## Height, Aspects 

```python
sns.displot(cars, x = 'weight',
            col='origin', height=2)
```

```python
sns.displot(cars, x = 'weight',
            col='origin', height=3, aspect=1.8)
```

## Other Keywords: Plot Specifics 

```python
sns.displot(cars, x = 'weight',bins = 20)
```

```python
sns.displot(cars, x = 'weight', kind='kde', fill=True, lw=3)
```

### rug_kws

```python
sns.displot(cars, x = 'weight', kind='kde', rug=True, rug_kws={'height':0.05})
```

## FacetGrid Styling 

```python
sns.displot(cars, x = 'weight', col='origin', 
            facet_kws={'despine':False})
```

```python
g = sns.displot(x='weight', data=cars, col='origin')

for col_val, ax in g.axes_dict.items():
    if col_val == "europe":
        ax.set_facecolor("#E2F2FF")
```

## Examples, Thumbnail

```python
plt.rcParams['font.size'] = 14
#This code is using the matplotlib library to change the default font size for all elements on a plot. 
# The plt.rcParams is a dictionary that stores the configuration settings for matplotlib.
sns.displot(x='weight', y='horsepower', data=cars)
plt.xlabel('Weight', fontsize=16)
plt.ylabel('Height', fontsize=16);
```

```python
g = sns.displot(x='weight', row='origin', hue='origin', palette='crest', 
            data=cars, height=4, aspect=2)
g.set_titles(row_template='{row_name} cars')
plt.xlabel("Weight")
plt.xticks([2000, 3000, 4000, 5000]);
```

```python
g = sns.displot(x='weight' ,
                color='#8711ac', 
                data=cars, height=4, aspect=0.8, kde=True)
g.set_titles(col_template='')
g.set_xlabels('')
g.set_ylabels('')
g.set_xticklabels([])
g.set_yticklabels([])
plt.xlim(1500, 5500)
plt.ylim(0, 80)
```

```python
g = sns.displot(x='weight', y='horsepower', row='origin', color='#48804D',
                data=cars[cars.origin.isin(['europe','japan'])], 
                height=3, aspect=2.3, row_order=['europe', 'japan'],
                facet_kws={'despine': False})
g.set_titles(row_template='')
g.set_xlabels('')
g.set_ylabels('')
g.set_xticklabels([])
g.set_yticklabels([])
plt.tight_layout();
```
