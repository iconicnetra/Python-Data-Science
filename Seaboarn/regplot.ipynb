# Example Using Different Parameters

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(0)
x = np.random.normal(size=100)
y = 2 * x + np.random.normal(size=100)

# Create DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Create a regplot
sns.regplot(x='X', y='Y', data=data, scatter=True, fit_reg=True, ci=95, n_boot=1000, order=1, logx=False,
            scatter_kws={'color': 'blue', 'alpha': 0.5}, line_kws={'color': 'red'}, ax=None)

# Add title and labels
plt.title('Regression Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()
```

# Shwoing Equation in the Line 

```python
import numpy as np
```

```python
#Creating a Regplot 
ax = sns.regplot(diamonds, x = 'carat', y = 'price', fit_reg=True, scatter=True, ci = None)

#Getting Regression Coefficients 
slope, intercept = np.polyfit(x, y, 1)

#Adding Equation to the Plot 
x_text = np.min(x) + (np.max(x) - np.min(x)) * 0.8
y_text = np.max(y) - (np.max(y) - np.max(y)) * 0.1
ax.text(x_text, y_text, f'y = {slope:.2f}x + {intercept:.2f}', fontsize = 14, color = 'red')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(0)
x = np.random.normal(size=100)
y = 2 * x + np.random.normal(size=100)

# Create DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Create a regplot
ax = sns.regplot(x='X', y='Y', data=data, scatter=True, fit_reg=True, ci=None)

# Get regression line coefficients
slope, intercept = np.polyfit(x, y, 1)

# Add equation to the plot
x_text = np.min(x) + (np.max(x) - np.min(x)) * 0.5
y_text = np.max(y) - (np.max(y) - np.min(y)) * 0.1
ax.text(x_text, y_text, f'y = {slope:.2f}x + {intercept:.2f}', fontsize=12, color='red')

# Add title and labels
plt.title('Regression Plot with Equation')
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()
```

#### Robust Regression

```python
x_example=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y_example=[0.1, 0.8, 2.2, 2.7, 3.8, 4.5, 6.2, 6.8, 7.9, 9.4, 30.4]
```

```python
sns.regplot(x=x_example, 
            y=y_example,
            ci=None
            );
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size and aspect ratio
plt.figure(figsize=(8, 6))

# Plot the data and the regression model
sns.regplot(x=x_example, 
            y=y_example,
            ci=None,
            robust=True,
            # Change the color and marker of the points
            color="green",
            marker="+"
            )

# Add labels and title
plt.xlabel("X variable")
plt.ylabel("Y variable")
plt.title("Robust linear regression with seaborn")

# Show the plot
plt.show()
```

```python
p = sns.lmplot(x='carat', y='price', 
               data=diamonds[diamonds.color.isin(['E', 'J'])], 
               hue='color', 
               order=2,
               palette=[green, orange])
plt.xlabel('Carat', fontsize=18)
plt.ylabel('Price', fontsize=18)
p._legend.remove()
plt.legend(fontsize=16)
plt.xticks([])
plt.yticks([])
plt.tight_layout();
```

```python
sns.jointplot(x='carat', y='price', data=diamonds, 
              kind='reg', 
              color='purple')
plt.xlabel('')
plt.ylabel('')
plt.xlim(-0.1, None)
plt.ylim(-2000, None)
plt.tight_layout();
```
