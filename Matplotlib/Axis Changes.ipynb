```python
import matplotlib.pyplot as plt 
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [21, 22, 26, 28, 33, 41]

plt.plot(x, y)

plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [21, 22, 26, 28, 33, 41]

plt.plot(x, y, color = 'green', marker = 'o')

plt.show()
```

```python
#Changes According to X-axis 

x = [1, 2, 3, 4, 5, 6]
y = [21, 22, 26, 28, 33, 41]

plt.plot(x, y)

plt.xticks(x)

plt.show()
```

```python
#Changes According to X-axis 

x = [1, 2, 3, 4, 5, 6]
y = [21, 22, 26, 28, 33, 41]

plt.plot(x, y)

plt.xticks(x)
plt.yticks(x)

plt.show()
```

```python
#Changes according to labels 

#Changes According to X-axis 

x = [1, 2, 3, 4, 5, 6]
y = [21, 22, 26, 28, 33, 41]

plt.plot(x, y)

plt.xticks(x, labels=["A", "B", "C", "D", "E", "F"])

plt.show()
```

```python
#Working with limits 
x = [1, 2, 3, 4, 5, 6]
y = [21, 22, 26, 28, 33, 41]

plt.plot(x, y)

plt.xlim(0, 10)
plt.ylim(20, 45)

plt.show()
```

```python
#Working with limits 
x = [1, 2, 3, 4, 5, 6]
y = [21, 22, 26, 28, 33, 41]

plt.plot(x, y, color = 'red', marker = 'o')

plt.xlim(0, 10)
plt.ylim(20, 45)

plt.show()
```
