## Working with Different Options

```python
import numpy as np
```

```python
j = sns.jointplot(x='waiting', y='duration', 
                  data=geyser.iloc[:, :-1]+np.random.normal(0, 0.15, size=geyser.iloc[:, :-1].shape), #adding "jitter"
                  color='white')
j.plot_marginals(sns.rugplot, height=.5, color='green')
j.ax_joint.set_xlabel('')
j.ax_joint.set_ylabel('')
```

```python
j = sns.jointplot(x='waiting', y='duration', data=geyser, color='green', kind='kde')
j.ax_joint.set_xlabel('')
j.ax_joint.set_ylabel('');
```

```python
j = sns.jointplot(x='waiting', y='duration', 
                  data=geyser.iloc[:, :-1]+np.random.normal(0, 0.15, size=geyser.iloc[:, :-1].shape), #adding "jitter"
                  color= 'green')
j.plot_marginals(sns.rugplot, height=-.15, color='black', clip_on=False)
j.plot_joint(sns.kdeplot, color='black', levels=5)
j.ax_joint.set_xlabel('')
j.ax_joint.set_ylabel('')
j.ax_joint.set_ylim(0.5, 6)
j.ax_joint.set_xlim(30, 110);
```

## Plot Kinds

There are currently six kinds of jointplots allowed in seaborn:
1. scatter (default)
2. kde
3. reg 
4. hist
5. hex
6. resid

### Scatter (Default)

```python
sns.jointplot(geyser, x = 'waiting', y = 'duration',
              kind='scatter')
```

### KDE

```python
sns.jointplot(geyser, x = 'waiting', y = 'duration',
              kind='kde')
```

### Reg

```python
sns.jointplot(geyser, x = 'waiting', y = 'duration',
              kind='reg')
```

### Histogram

```python
sns.jointplot(geyser, x = 'waiting', y = 'duration',
              kind='hist')
```

```python
sns.jointplot(geyser, x = 'waiting', y = 'duration',
              kind='hex')
```

```python
sns.jointplot(geyser, x = 'waiting', y = 'duration',
              kind='resid')
```
