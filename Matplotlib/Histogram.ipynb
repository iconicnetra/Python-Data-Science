```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

## Creating a Simple Histogram

```python
plt.style.use("fivethirtyeight")

ages = [11, 22, 33, 44, 23, 34, 54, 65, 67, 55, 44, 35,54, 66, 56,34,29]

plt.hist(ages, bins=8)

plt.title("Historgram of Ages of the Respondents")
plt.xlabel("Ages")
plt.ylabel("Frequency")

plt.tight_layout()

plt.show()
```

## Using DATA FRAME

```python
df = pd.read_csv("E:\Python\Python Data Science\Matplotlib\data.csv")
df.head()
```

```python
#Creating Histogram of Matks Mathematics

plt.style.use("ggplot")

plt.hist(df['Marks_Math'], bins = 10, color='g')

plt.xlabel("Marks in Mathematics")
plt.ylabel("Frequency")
plt.title("Distribution of Marks in Mathematics")

plt.show()
```

```python
#Next Method

plt.style.use("ggplot")
cs_marks = df['CS_Marks']

plt.hist(cs_marks, bins = 10,
         edgecolor = 'black', color = 'green')

plt.xlabel("Marks in Computer Science")
plt.ylabel("Frequency")
plt.title("Distribution of Marks in CS")

plt.legend()

plt.show()
```

### Using Bins as an Interval Data

```python
bins = [20, 25, 30, 35, 40, 45, 50]

plt.hist(cs_marks, bins = bins,
         edgecolor = 'black', color = 'y')

plt.xlabel("Marks in CS")
plt.ylabel("Frequency")
plt.title("Distribution of Marks in CS")

plt.legend()
plt.grid(True)

plt.show()
```

```python
#Next Method

plt.style.use("ggplot")
cs_marks = df['CS_Marks']

plt.hist(cs_marks, bins = 10,
         edgecolor = 'black', color = 'green', histtype='step')

plt.xlabel("Marks in Computer Science")
plt.ylabel("Frequency")
plt.title("Distribution of Marks in CS")

plt.legend()

plt.show()
```

```python
median = df['CS_Marks'].median()
median
```

```python
#Adding Median in the Data

plt.style.use("ggplot")
cs_marks = df['CS_Marks']

median = 41

plt.hist(cs_marks, bins = 10,
         edgecolor = 'black', color = 'green', align='mid')

plt.axvline(median, color = 'red', label = 'Median Marks', linewidth = 1.5)

plt.xlabel("Marks in Computer Science")
plt.ylabel("Frequency")
plt.title("Distribution of Marks in CS")

plt.legend()

plt.show()
```
