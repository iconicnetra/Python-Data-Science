```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.figure(figsize=(3,3), dpi=100)


plt.plot(x, y, 'ro', label = "$x^2$", linestyle="dashed", linewidth=2, markersize=5)

plt.title("My first Graph") 
plt.xlabel("x-values")
plt.ylabel("y-values")

plt.grid(True)
plt.legend()

plt.show()

plt.savefig("Graph of $x^2$.png")
```
