# Seaborn Catplot (Categorical Plot)

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

## Loading Dataset 

```python
#loading and dropping the missings 
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
#Filtering to 4, 6, 8 - cylinders cars
cars = cars[cars.cylinders.isin([4, 6, 8])]
cars.head()
```

```python
cars.shape
```

```python
#Add old/new to type column
cars['type'] = ['old' if x <= 76 else 'new' for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

```python
sns.set_style("darkgrid")
```

```python
g = sns.catplot(cars, x = 'cylinders', y = 'horsepower',
                row='origin',
                kind='box',
                aspect=1.5,
                height=4)
g.set_titles(row_template="{row_name} cars")
```

## Thumbnail

```python
g = sns.catplot(cars, x = 'cylinders', y = 'mpg',
                col='type',
                kind='violin',
                aspect=1.8,
                height=5,
                facet_kws={'ylim':(0, 50)},
                palette=['lightgray', '#B58389', '#7C2021'])
g.set_titles(col_template="{col_name} cars")
plt.tight_layout()
```

```python
g = sns.catplot(x='horsepower', y=['']*len(cars),
                hue='cylinders',
                data=cars,
                kind='strip',
                aspect=4.8,
                height=3,
                s=10,
                legend=False,
                palette=['lightgray', '#B58389', '#7C2021'],
               )
g.set_titles(col_template="{col_name} cars")
plt.tight_layout();
```

## Small Multiple (FacetGrid)
