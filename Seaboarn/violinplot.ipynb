```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("dark")
```

# Violin Plot 

#### Loading Data Set 

```python
cars = sns.load_dataset("mpg").dropna()
```

```python
cars.head()
```

```python
cars.info()
```

```python
cars.describe()
```

```python
cars.shape
```

```python
cars.cylinders.value_counts()
```

```python
#Creating Plot 
sns.set_style("whitegrid")
sns.violinplot(cars, x = "displacement")
```

#### Creating Multiple Plots 

```python
sns.violinplot(cars, x = "cylinders", y = "displacement")
```

```python
sns.violinplot(cars, x = "cylinders", y = "displacement",
               hue = "cylinders")
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
cars.head()
```

```python
sns.violinplot(cars, 
               x = "cylinders", y = "displacement", hue = "cylinders")
```

```python
sns.violinplot(cars, 
               x = "origin", y = "displacement", hue = "cylinders")
```

### ViolinPlot Options

```python
sns.violinplot(data = cars[cars.origin.isin(["japan", "europe"])], 
               x = "cylinders", y = "displacement", hue = "origin")

plt.legend(loc=2)
```

### Splitting 

```python
sns.violinplot(data=cars[cars.origin.isin(["japan", "europe"])], hue = "origin",
               x = "cylinders", y = "displacement", 
               split=True)

plt.legend(loc = 2) #place of the legend
```

#### Showing Inner Values 

```python
sns.violinplot(data=cars[cars.origin.isin(["japan", "europe"])], hue = "origin",
               x = "cylinders", y = "displacement", 
               split=True, inner='quartiles')

plt.legend(loc = 2)
```

```python
cars[cars.origin.isin(["japan", "europe"])].describe()
```

### Scalling 

```python
sns.violinplot(data=cars[cars.origin.isin(["japan", "europe"])], hue = "origin",
               x = "cylinders", y = "displacement", 
               split=True, 
               scale='count') #density_norm='count' is new term for new seaborn library

plt.legend(loc = 2)
```

```python
sns.violinplot(data=cars[cars.origin.isin(["japan", "europe"])], hue = "origin",
               x = "cylinders", y = "displacement", 
               split=True, 
               density_norm='count') #density_norm='count' is new term for new seaborn library

plt.legend(loc = 2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               density_norm='count',
               common_norm=False
              ) #new_name for scale_hue = common_norm

plt.legend(loc=2);
```

```python
cars[cars.origin.isin(["japan", "europe"])].groupby('cylinders').origin.value_counts()
```

### Styling

```python
sns.violinplot(cars, x = "origin", y = "displacement",
               order=['japan', 'europe', 'usa'])
```

```python
sns.violinplot(cars, x = "origin", y = "displacement",
               linewidth=3)
plt.xlabel("Origin")
plt.ylabel("Displacement")
```

#### KDE Options

```python
sns.violinplot(cars, x = "origin", y = "displacement",
               linewidth=3, bw_adjust=0.9, hue = 'origin') #Changing the bw = BandWidth
plt.xlabel("Origin")
plt.ylabel("Displacement")
```
